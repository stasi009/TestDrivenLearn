```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('../data/users.dat', sep='::', header=None,names=unames)

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('../data/ratings.dat', sep='::', header=None,names=rnames)

# below just shows 'read_csv' is compatible with 'read_table'
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_csv('../data/movies.dat', sep='::', header=None,names=mnames)
```

```python
users.set_index("user_id",inplace=True)
movies.set_index("movie_id",inplace=True)
```

```python
temp = pd.merge(users,ratings,left_index=True,right_on="user_id")
full_df = pd.merge(temp,movies,left_on="movie_id",right_index=True)
```

```python
full_df = full_df.loc[:,["user_id","gender","age","occupation","zip","movie_id","title","genres","rating","timestamp"]]
```

```python
full_df.pivot_table("rating",index="title",columns="gender",aggfunc="mean")
```
