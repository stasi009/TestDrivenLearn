<a id="serial_correlation"></a>
### Serial Correlation
serial correlation, which is the correlation between each element and the next (or the previous). To compute serial correlation, we can shift a signal and then compute the correlation of the shifted version with the original.  
**here, we just considering a fixed shift. when the shift is varied, we calculate the serial-correlation against all possible shift, then the result becomes [Autocorrelation](#autocorrelation)**

we talk about **"correlated signal"** and **"uncorrelated signal"**:
* correlated signal: each value depends on the preceding value
* uncorrelated signal: means that, in the signal, each value is independent of the others

```python
def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr
```

```python
signal = thinkdsp.UncorrelatedGaussianNoise()
wave = signal.make_wave(duration=0.5, framerate=11025)
serial_corr(wave)
```

the serial-correlation is very small, which is just what we expected (because the signal is a uncorrelated Gaussian noise). 

<a id="autocorrelation"></a>
### Autocorrelation

<span style="color:red;font-size:1.5em;font-weight:bold">autocorrelation “is intended for time series that do not exhibit a trend and have zero mean”</span>

You can think of serial_corr as a function that maps from each value of lag to the corresponding correlation, and we can calculate Autocorrelation by looping through differnt values of lag.

```python
def autocorr(wave):
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs
```

<a id="better_estimate_freq"></a>
#### <span style="color:darkorange">use Autocorrelation to better estimate frequency</span>

the problem we are facing is: giving a piece of sound, we want to know the frequency of **a particular time**. the difficulty is how to choose the segment length:
* <span style="color:red">since we want to know the frequency of a "time point" (a particular time), then the segment length must be small, but a small segment has poor frequency resolution</span>
* <span style="color:red">to get better frequency resolution, we can use a longer segment. but longer segment violates our requirement, that we want to know the frequency of a "particular time point"</span>

mathmatically, if we choose a period of "p" seconds from the original second. the original sound is sampled at "r" Hz. then, the chosen period has "pr" points. then the **frequency resolution is r/(pr)=1/p**. so larger period, poor time resolution, but better frequency resolution.

for example, for a sound "voicedownbew.wav", we want to know the frequency at 0.2 seconds

```python
wave = thinkdsp.read_wave('voicedownbew.wav')
wave.normalize()

duration = 0.01 # frequency resolution is 1/0.01=100hz
segment = wave.segment(start=0.2, duration=duration)
spectrum = segment.make_spectrum()
thinkplot.preplot(cols=2)

thinkplot.subplot(1)
segment.plot()
thinkplot.config(title="time domain",xlabel="time (s)")

thinkplot.subplot(2)
spectrum.plot(high=16,marker="o",mfc="r",color="b")
thinkplot.config(title="frequency domain",xlabel='frequency (Hz)', ylabel='amplitude')
```

from above plot, we can conclude the signal frequency is closer to 400Hz. but <span style="color:orangered;font-weight:bold;">since the frequency resolution is 100Hz, so that "closeness" is compared with 300hz and 500hz. will be any "closer" frequency within 300hz~500hz? we don't know.</span>

we can estimate the frequency more precisely by using autocorrelation. the basic idea is: <span style="color:orangered;font-weight:bold;">we expect the autocorrelation to spike when the lag equals the period.</span>

```python
lags, corrs = autocorr(segment)
thinkplot.plot(lags, corrs)
thinkplot.config(xlabel='lag (index)', ylabel='correlation', ylim=[-1, 1],title="Autocorrelation")
```

```python
# we found the peak is at 100 (0 means no shift, always be 1, doesn't count)
lags = xrange(99,102)
for lag in lags:
    period = float(lag) / segment.framerate
    frequency = 1 / period
    print "lag=%d, frequency = %3.2fHz"%(lag,frequency)
```

<span style="color:darkorange;font-weight:bold">we can see that, by using autocorrelation, So the frequency precision is less than 10 Hz (445.45hz-436.63hz=8.82hz), compared with 100 Hz using the spectrum.</span>
