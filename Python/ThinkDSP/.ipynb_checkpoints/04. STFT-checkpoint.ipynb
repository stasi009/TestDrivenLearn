<a id="gabor_limit"></a>
### The Gabor Limit

the sampling frequency is r samples per seconds. and the segments length is n.
* the time resolution is: n/r
* the frequency resolution is: r/n  

This tradeoff is called the **Gabor limit**

```python
signal = thinkdsp.Chirp(start=220, end=440)
wave = signal.make_wave(duration=1, framerate=12000)

segment_sizes = [512,1024,2048,4096]
thinkplot.preplot(rows=2,cols=2)

for index,segment_size in enumerate(segment_sizes):
    spectrogram = wave.make_spectrogram(seg_length=segment_size)
    thinkplot.subplot(index+1)
    spectrogram.plot(high=100)# high is the index of highest frequency you want to draw
    thinkplot.config(title="segment_size=%d"%(segment_size),ylim=(200,500),xlabel="time",ylabel="frequency")
```

above 4 plots are demonstration of the trade-off between time resolution and frequency resolution
1. the first plot, segment size is 512. small segment, has good time resolution, but poor frequency resolution. from the plot, we can distinguish **horizontal** color difference, but we cannot distinguish color difference along the **vertial** direction.
2. the third plot, segment size is 2048. large segment, has good frequency resolution, but poor time resolution. from the plot, we can distinguish **vertical** color difference, but we cannot distinguish color difference along the **horizontal** direction.
