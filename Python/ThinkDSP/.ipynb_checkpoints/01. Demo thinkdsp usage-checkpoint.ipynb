<a id="construct_signal"></a>
### construct a signal
* freq is frequency in **Hz**. 
* amp is amplitude in unspecified units where 1.0 is generally the largest amplitude we can play.
* offset is a phase offset in **radians**.

```python
cos_sig = thinkdsp.CosSignal(freq=440, amp=1.0, offset=0)
sin_sig = thinkdsp.SinSignal(freq=880, amp=0.5, offset=0)
mix_sig = sin_sig + cos_sig # operators have been overriden on signal object to do math
```

```python
mix_sig.plot()# underneath, it just construct a wave with default duration and sampling frequency and plot that wave
```

<a id="make_wave"></a>
### construct wave from signal

call signal.make_wave to construct a wave from signal
* duration is the length of the Wave in **seconds**. 
* start is the start time, also in **seconds**. 
* framerate is the (integer) number of samples per second, aka, sampling frequency.

```python
wave = mix_sig.make_wave(duration=0.5, start=0, framerate=1600)
wave.plot()
```

### Wave's ts and ys properties

```python
duration = sin_sig.period * 10
sample_freq = 16000
sin_wave = sin_sig.make_wave(start=0,duration=duration,framerate=sample_freq)
cos_wave = cos_sig.make_wave(start=0,duration=duration,framerate=sample_freq)
mix_wave = mix_sig.make_wave(start=0,duration=duration,framerate=sample_freq)
```

```python
plt.plot(sin_wave.ts,sin_wave.ys,label="sin")
plt.plot(cos_wave.ts,cos_wave.ys,label="cos")
plt.plot(mix_wave.ts,mix_wave.ys,label="mix")
plt.grid()
plt.legend(loc="best")
plt.xlabel("time (unit:s)")
plt.ylabel("value (unit:NA)")
```

<a id="make_spectrum"></a>
### Wave's make_spectrum

```python
spectrum = mix_wave.make_spectrum()
spectrum.plot()
```

Spectrum provides three methods that modify the spectrum:
* **low_pass** applies a low-pass filter, which means that components above a given cutoff frequency are attenuated (that is, reduced in magnitude) by a factor.
* **high_pass** applies a high-pass filter, which means that it attenuates components below the cutoff.
* **band_stop** attenuates components in the band of frequencies between two cutoffs.

```python
spectrum.low_pass(cutoff=600, factor=0.01)
spectrum.plot()
```

```python
# call make_wave to convert a Spectrum back to a Wave
spectrum.make_wave().plot()
```
