```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get("http://www.yellowpages.com/search?search_terms=restaurant&geo_location_terms=Pullman%2C+WA")
soup = BeautifulSoup(r.content)
```

```python
print soup.prettify()
```

```python
all_links = soup.findAll("a")
for index,link in enumerate( all_links ):
    # there are multiple ways to access attributes and text
    # link.text, link.get("href")
    print "[{}]: <a href='{}'>{}</a>".format(index+1,link.attrs["href"],link.get_text())
```

```python
info_divs = soup.findAll("div",{"class":"info"})
```

```python
for index, item in enumerate(info_divs):
    # I have to use u'...', otherwise, it will raise Encoding Error
    # print u"[{}]: {}".format(index+1,item.text)
    # print "[{}]: {}".format(index+1,item.text.encode("ascii","ignore"))
    tag = item.find("a",{"class":"business-name"})
    if tag is not None:
        print "{}: {}".format(index+1,tag.text)
```
